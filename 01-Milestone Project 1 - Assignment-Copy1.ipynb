```python
def tic_tac():
    tic_tac_start()
    z= play_check()
    
    global p1_in
    global p2_in
    p1_in = []
    p2_in = []
    count =0
    while z == True:
        tic_tac_disp(p1_in,p2_in)
        
        if count==9:
            print("It's a draw")
            print("Thank you! See you again")
            break
        
        count +=1
        p1_in.append(input_func('Player1'))
        tic_tac_disp(p1_in,p2_in)
        if win_check(p1_in) == True:
            print("Player1 wins!!!!!!")
            break
        
        if count==9:
            print("It's a draw")
            print("Thank you! See you again")
            break
        
        count +=1
        p2_in.append(input_func('Player2'))
        tic_tac_disp(p1_in,p2_in)
        if win_check(p2_in) == True:
            print("Player2 wins!!!!!!")
            break
    else:
        print("Thank you! See you again")
    
    
    
```

```python
from IPython.display import clear_output


def tic_tac_start():
    print("Welcome to Tic Tac Toe!\n")
    global pp11
    global pp22
    while True:
        p1 = input("Player1: Do you want to be X or O?\n")
        if p1.lower() == 'x':
            pp11,pp22 = "X","O"
            break
        elif p1.lower() == 'o':
            pp11,pp22 = "O","X"
            break
        else:
            continue
    
    print("Player1 will go first.")
    print()

    
def play_check():
    while True:
        ck = input("Are you read to play? Enter Yes or No\n")
        if ck.lower() == "no":
            print("Thank you! See you again")
            return False
        elif ck.lower()== 'yes':
            return True
        else:
            pass
    
def tic_tac_disp(lst1,lst2):
    clear_output()
    pd={'n1':' ','n2':' ','n3':' ','n4':' ','n5':' ','n6':' ','n7':' ','n8':' ','n9':' '}
    for item1 in lst1:
        pd['n'+item1]=pp11
    for item2 in lst2:
        pd['n'+item2]=pp22
    print('\t7\t8\t9\t\n\n\t4\t5\t6\t\n\n\t1\t2\t3\t')
    print('\n\n')
    print("\t     |     |     ")
    print(f"\t  {pd['n7']}  |  {pd['n8']}  |  {pd['n9']}  ")  
    print("\t_________________")
    print("\t     |     |     ")     
    print(f"\t  {pd['n4']}  |  {pd['n5']}  |  {pd['n6']}  ")
    print("\t_________________")
    print("\t     |     |     ")     
    print(f"\t  {pd['n1']}  |  {pd['n2']}  |  {pd['n3']}  ")
    
def input_func(player):
    while True:
        ins=input(f"{player}: Choose a number from 1-9: ")
        if (ins not in p1_in+p2_in) and (ins in list(map( lambda ch: str(ch),range(1,11)))):
            return ins
            
        else:
            continue
    

def win_check(win_lst):
    wins = ('123','456','789','147','258','369','159','357')

    win1=[]
    for item in wins:
        for ch in win_lst:
            item = item.replace(ch,"x")

        win1.append(item)
    return 'xxx' in win1
```
